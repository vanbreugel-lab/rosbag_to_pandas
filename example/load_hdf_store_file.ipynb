```python
import pandas as pd
```

# Filename for HDF5 create by bag2hdf5

Use the "pandas" one, which is an HDF Store file. 

```python
filename = 'drone_flight_demo_20260130_pandas.hdf5'
```

# Load repackaged data with pandas

```python
with pd.HDFStore(filename, 'r') as store:
    keys = store.keys()
    print(keys)
```

```python
# Load data from a specific key
df = pd.read_hdf(filename, keys[0])
```
