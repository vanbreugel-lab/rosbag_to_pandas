# Filename for HDF5 create by bag2hdf5

```python
filename = 'drone_flight_demo_20260130.hdf5'
```

# List keys

```python
rosbag_to_pandas.find_all_datasets(filename)
```

# Load a specific key

```python
df_mavros = rosbag_to_pandas.get_pandas_dataframe_from_uncooperative_hdf5(filename, key='mavros/imu/data')
df_mavros
```

# Repackage the raw hdf5 into a pandas HDFStore file

```python
output_filename = filename.split('.')[0] + '_pandas.hdf5'
```

```python
all_dfs = rosbag_to_pandas.save_all_datasets_to_hdf5(filename, output_filename)
```

# Load repackaged data with pandas

```python
# Use a helper function to look at what keys are available.
keys = rosbag_to_pandas.list_hdf5_keys(output_filename)
print(keys)
```

```python
# Load data from a specific key
df = pd.read_hdf(output_filename, keys[0])
```
